```python



non_numeric_cols = ["Country", "ISO", "Series Name", "Series Code"]
num_cols = [col for col in df.columns if col not in non_numeric_cols]


for col in num_cols:

    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.strip()
    )


    df[col] = pd.to_numeric(df[col], errors="coerce")



# impossible values
invalid_masks = {}

for col in num_cols:
    mask = (df[col] < 0) | (df[col] > 1e12)  # adjust threshold as needed
    if mask.sum() > 0:
        invalid_masks[col] = mask



for col, mask in invalid_masks.items():
    print(f"Column '{col}' has {mask.sum()} suspicious values:")
    print(df.loc[mask, [col]].head())


#  extreme values

for col, mask in invalid_masks.items():
    df.loc[mask, col] = np.nan

```

```python
year_cols = [str(y) for y in range(2000, 2024)]
for col in year_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')


df[year_cols] = df[year_cols].fillna(df[year_cols].median())
```
